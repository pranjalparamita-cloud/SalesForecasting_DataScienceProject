```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True) #Converting date front
```

```python
df['Year'] = df['Date'].dt.year   #taking year from date
df['Month'] = df['Date'].dt.month    #taking month from date
df['Week'] = df['Date'].dt.isocalendar().week   #taking week
```

```python
df = df.drop('Date', axis=1)      #removing date column
```

```python
df.head() 
```

```python
x = df.drop('Weekly_Sales', axis=1)    #input data
y = df['Weekly_Sales']    #output data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)   #Splitting data
```

```python
from sklearn.ensemble import RandomForestRegressor  #importing RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(x_train, y_train)      #model training
```

```python
y_pred = model.predict(x_test)  #making prediction
```

```python
from sklearn.metrics import mean_absolute_error, r2_score   # import to find error and r2 score
```

```python
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("MAE:", mae)
print("R2 Score:", r2)
```

```python
accuracy = r2 * 100
print("Model Accuracy:", accuracy, "%")  #finding accuracy
```
